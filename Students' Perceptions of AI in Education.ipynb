```python
from sklearn.preprocessing import OneHotEncoder
categorical_columns = data.select_dtypes(exclude=['number']).columns
encoded_data = pd.get_dummies(data[categorical_columns])
all_numeric_data = pd.concat([data[numeric_columns], encoded_data], axis=1)
correlation = all_numeric_data.corr()
```

```python
#Q1 – AI Knowledge Level Histogram
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data["Q1.AI_knowledge"], bins=10, kde=True, color="skyblue")
plt.title("Q1: Self-reported AI Knowledge Level")
plt.xlabel("Knowledge Level (1-10)")
plt.ylabel("Number of Students")
plt.show()
```

```python
#Q2 – AI Information Sources
```

```python
sources_cols = ['Q2#1.Internet', 'Q2#2.Books/Papers', 'Q2#3.Social_media', 'Q2#4.Discussions', 'Q2#5.NotInformed']
sources_labels = ['Internet', 'Books/Papers', 'Social Media', 'Discussions', 'Not Informed']
sources_counts = data[sources_cols].sum()

df = pd.DataFrame({
    'source': sources_labels,
    'count': sources_counts.values
})

plt.figure(figsize=(10, 6))
sns.barplot(x='source', y='count', hue='source', data=df, palette="viridis")
plt.title("Q2: Sources of AI Information")
plt.ylabel("Number of Mentions")
plt.xticks(rotation=30)
plt.show()
```

```python
#Q3 – Agreement on AI’s Role in Society
```

```python
q3_cols = ['Q3#1.AI_dehumanization','Q3#2.Job_replacement','Q3#3.Problem_solving', 
    'Q3#4.AI_rulling_society']
q3_labels = ['Dehumanization', 'Job Replacement', 'Problem Solving', 'AI Ruling Society']

q3_data = pd.DataFrame()
for i, col in enumerate(q3_cols):
    q3_data[q3_labels[i]] = data[col].value_counts().sort_index()

q3_data = q3_data.T.fillna(0)

q3_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='Set1')
plt.title("Q3: Opinions on AI's Role in Society")
plt.ylabel("Number of Students")
plt.xlabel("Statements")
plt.xticks(rotation=30)
plt.legend(title="Agreement Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Q4 – Economic Impact of AI
```

```python
q4_cols = [
    'Q4#1.AI_costly', 
    'Q4#2.Economic_crisis', 
    'Q4#3.Economic_growth', 
    'Q4#4.Job_loss'
]
q4_labels = ['Costly AI', 'Global Crisis', 'Economic Growth', 'Job Loss']

q4_data = pd.DataFrame()
for i, col in enumerate(q4_cols):
    q4_data[q4_labels[i]] = data[col].value_counts().sort_index()

q4_data = q4_data.T.fillna(0)

q4_data.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='coolwarm')
plt.title("Q4: Perceived Economic Effects of AI")
plt.ylabel("Number of Students")
plt.xlabel("Statements")
plt.xticks(rotation=30)
plt.legend(title="Agreement Level", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
#Q5 – Emotions Toward AI
```

```python
emotion_map = {1: "Curiosity", 2: "Fear", 3: "Indifference", 4: "Trust"}
emotion_counts = data["Q5.Feelings"].map(emotion_map).value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(x=emotion_counts.index, y=emotion_counts.values, hue=emotion_counts.index, legend=False, palette="pastel")
plt.title("Q5: Student Emotions Towards AI")
plt.ylabel("Number of Students")
plt.xlabel("Emotion")
plt.show()
```

```python
#Q6 – Perceived Areas of AI Impact
```

```python
impact_cols = [col for col in data.columns if col.startswith("Q6#")]
impact_labels = [col.split('#')[1] for col in impact_cols]
impact_counts = data[impact_cols].sum()

plt.figure(figsize=(10, 6))
sns.barplot(x=impact_labels, y=impact_counts.values, hue=impact_labels, legend=False, palette="cubehelix")
plt.title("Q6: Perceived Areas of AI Impact")
plt.xticks(rotation=45)
plt.ylabel("Number of Mentions")
plt.show()
```
