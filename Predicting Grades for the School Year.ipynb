```python
#Relationship Analysis for corelations between coefficients
```

```python
student_mat_encoded = student_mat.copy()
student_mat_encoded['sex'] = student_mat_encoded['sex'].map({'F': 0, 'M': 1})
correlation = student_mat_encoded.corr()
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns,annot=True)
```
